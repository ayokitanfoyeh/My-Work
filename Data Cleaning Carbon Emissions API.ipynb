```python
import json 

# Define an empty list to store the itineraries
itineraries = []

# Iterate over the rows of the dataframe
for _, row in itinerary_df.iterrows():
    
    # Define a list to store the locations
    locations = []
    
    # Iterate over the columns that contain the locations
    for col in ['Location1', 'Location2','Location3', 'Location4','Location5', 
              'Location6','Location7', 'Location8',
              'Location9', 'Location10'
             ,'Location11', 'Location12',
              'Location13', 'Location14']:
        # Check if the location is not "SKI" or empty
        if row[col] not in ["SKI", ""] and not pd.isnull(row[col]):
            locations.append(row[col])
    
    # Iterate over the pairs of locations to create the itineraries
    for i in range(len(locations) - 1):
        itinerary = {
            "Departure": locations[i],
            "Arrival": locations[i + 1],
            "TravelClass": row["TravelClass"]}
        itineraries.append(itinerary)

# Create the final JSON object
json_obj = {"Itineraries": itineraries}

# Convert the JSON object to a string with no indentation and print it
json_str = json.dumps(json_obj, indent=4)
json_str = json_str.replace('"Departure":', 'Departure:')
json_str = json_str.replace('"Arrival":', 'Arrival:')
json_str = json_str.replace('"TravelClass":', 'TravelClass:')
json_str = json_str.replace('"Itineraries":', 'Itineraries:')
print(json_str)
```
